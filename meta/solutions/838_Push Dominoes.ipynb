# 838. Push Dominoes

There are n dominoes in a line, and we place each domino vertically upright. In the beginning, we simultaneously push some of the dominoes either to the left or to the right.After each second, each domino that is falling to the left pushes the adjacent domino on the left. Similarly, the dominoes falling to the right push their adjacent dominoes standing on the right.When a vertical domino has dominoes falling on it from both sides, it stays still due to the balance of the forces.For the purposes of this question, we will consider that a falling domino expends no additional force to a falling or already fallen domino.You are given a string dominoes representing the initial state where:dominoes[i] = 'L', if the ith domino has been pushed to the left,dominoes[i] = 'R', if the ith domino has been pushed to the right, anddominoes[i] = '.', if the ith domino has not been pushed.Return a string representing the final state. **Example 1:**Input: dominoes = "RR.L"Output: "RR.L"Explanation: The first domino expends no additional force on the second domino.**Example 2:**Input: dominoes = ".L.R...LR..L.."Output: "LL.RR.LLRRLL.." **Constraints:**n == dominoes.length1 <= n <= 105dominoes[i] is either 'L', 'R', or '.'.

## Solution Explanation
To solve this problem, we need to simulate the falling dominoes and determine their final state. The key insight is to understand how forces propagate through the dominoes.My approach uses a two-pass algorithm:1. First, we'll process forces coming from the right (R) and propagate them to the right.2. Then, we'll process forces coming from the left (L) and propagate them to the left.For each domino position, we'll calculate:* The distance to the nearest 'R' on its left (or infinity if none exists)* The distance to the nearest 'L' on its right (or infinity if none exists)Then, for each position:* If the distance to 'R' is less than the distance to 'L', the domino falls right* If the distance to 'L' is less than the distance to 'R', the domino falls left* If the distances are equal, the domino remains standing* If both distances are infinity, the domino remains in its original stateHowever, there's a more efficient approach using a simulation with forces. We can:1. Iterate through the dominoes2. For each 'R', apply a rightward force until we hit an 'L' or another 'R'3. For each 'L', apply a leftward force until we hit an 'R' or another 'L'4. When forces meet, determine the outcome based on their relative strengths

```python
def pushDominoes(dominoes: str) -> str:    # Add boundary dominoes to simplify logic    dominoes = 'L' + dominoes + 'R'    n = len(dominoes)    result = list(dominoes)        i = 0    while i < n:        # Skip already falling dominoes        if dominoes[i] == '.':            i += 1            continue                # Find the next non-vertical domino        j = i + 1        while j < n and dominoes[j] == '.':            j += 1                # Process the segment between i and j        left_force = dominoes[i]        right_force = dominoes[j] if j < n else None                # Apply forces based on the boundary dominoes        if left_force == right_force:  # Both 'L' or both 'R'            for k in range(i+1, j):                result[k] = left_force        elif left_force == 'R' and right_force == 'L':  # Opposing forces            # Calculate middle point            mid = (j - i - 1) // 2            for k in range(1, mid + 1):                result[i + k] = 'R'                result[j - k] = 'L'            # If odd number of dots, middle one stays vertical        # For 'L...R', dots remain vertical, no change needed                # Move to the next segment        i = j        # Remove the boundary dominoes we added    return ''.join(result[1:-1])
```

## Time and Space Complexity
* *Time Complexity**: O(n), where n is the length of the input string. We iterate through the string once, and for each segment between two non-vertical dominoes, we process each position at most once.* *Space Complexity**: O(n) for the result array that stores the final state of the dominoes. We also use a few additional variables that take O(1) space.
