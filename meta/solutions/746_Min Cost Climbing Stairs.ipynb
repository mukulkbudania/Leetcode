# 746. Min Cost Climbing Stairs

You are given an integer array cost where cost[i] is the cost of ith step on a staircase. Once you pay the cost, you can either climb one or two steps.You can either start from the step with index 0, or the step with index 1.Return the minimum cost to reach the top of the floor. **Example 1:**Input: cost = [10,15,20]Output: 15Explanation: You will start at index 1.- Pay 15 and climb two steps to reach the top.The total cost is 15.**Example 2:**Input: cost = [1,100,1,1,1,100,1,1,100,1]Output: 6Explanation: You will start at index 0.- Pay 1 and climb two steps to reach index 2.- Pay 1 and climb two steps to reach index 4.- Pay 1 and climb two steps to reach index 6.- Pay 1 and climb one step to reach index 7.- Pay 1 and climb two steps to reach index 9.- Pay 1 and climb one step to reach the top.The total cost is 6. **Constraints:**2 <= cost.length <= 10000 <= cost[i] <= 999

## Solution Explanation
This problem is a classic dynamic programming problem where we need to find the minimum cost path to reach the top of the stairs.The key insight is that at each step, we have two choices: we can either come from one step behind or two steps behind. We want to choose the path that minimizes the total cost.Let's define `dp[i]` as the minimum cost to reach the i-th step. Then:* `dp[0] = cost[0]` (starting from the first step)* `dp[1] = cost[1]` (starting from the second step)* For i ≥ 2: `dp[i] = cost[i] + min(dp[i-1], dp[i-2])` (we take the minimum of coming from one or two steps behind)To reach the top of the floor (which is beyond the last step), we can either come from the last step or the second-to-last step. So the final answer is `min(dp[n-1], dp[n-2])`, where n is the length of the cost array.We can optimize the space complexity by only keeping track of the last two values instead of the entire dp array.

```python
def minCostClimbingStairs(cost):    n = len(cost)        # Base cases    if n == 0:        return 0    if n == 1:        return cost[0]        # Initialize dp array with the first two steps    dp = [0] * n    dp[0] = cost[0]    dp[1] = cost[1]        # Fill the dp array    for i in range(2, n):        dp[i] = cost[i] + min(dp[i-1], dp[i-2])        # The final answer is the minimum cost to reach the top    # which can be reached from either the last or second-to-last step    return min(dp[n-1], dp[n-2])# Space-optimized versiondef minCostClimbingStairs_optimized(cost):    n = len(cost)        if n == 0:        return 0    if n == 1:        return cost[0]        # We only need to keep track of the last two values    prev1 = cost[0]    prev2 = cost[1]        for i in range(2, n):        current = cost[i] + min(prev1, prev2)        prev1, prev2 = prev2, current        # Return the minimum of the last two steps    return min(prev1, prev2)
```

## Time and Space Complexity
* *Time Complexity**: O(n), where n is the length of the cost array. We iterate through the array once, performing constant-time operations at each step.* *Space Complexity**: * For the first implementation: O(n) for the dp array.* For the optimized implementation: O(1) as we only use two variables regardless of the input size.
