Implement pow(x, n), which calculates x raised to the power n (i.e., xn). **Example 1:**Input: x = 2.00000, n = 10Output: 1024.00000**Example 2:**Input: x = 2.10000, n = 3Output: 9.26100**Example 3:**Input: x = 2.00000, n = -2Output: 0.25000Explanation: 2-2 = 1/22 = 1/4 = 0.25 **Constraints:**-100.0 < x < 100.0-231 <= n <= 231-1n is an integer.Either x is not zero or n > 0.-104 <= xn <= 104

## Solution Explanation
To implement the `pow(x, n)` function efficiently, we'll use the fast power algorithm (also known as binary exponentiation). This approach is much more efficient than the naive approach of multiplying x by itself n times.The key insight is that we can break down the exponentiation into smaller subproblems:* If n is even, x^n = (x^(n/2))^2* If n is odd, x^n = x * (x^(n/2))^2For negative exponents, we can use the property that x^(-n) = 1/(x^n).The algorithm works recursively or iteratively by repeatedly squaring the base and halving the exponent. This reduces the time complexity from O(n) to O(log n).

```python
def myPow(x: float, n: int) -> float:    # Handle edge case    if n == 0:        return 1.0        # Handle negative exponent    if n < 0:        return 1 / myPow(x, -n)        # Use binary exponentiation (fast power algorithm)    if n % 2 == 0:        # If n is even, x^n = (x^(n/2))^2        half_pow = myPow(x, n // 2)        return half_pow * half_pow    else:        # If n is odd, x^n = x * (x^(n/2))^2        half_pow = myPow(x, n // 2)        return x * half_pow * half_powWe can also implement this iteratively to avoid stack overflow for large values of n:def myPow(x: float, n: int) -> float:    # Handle edge case    if n == 0:        return 1.0        # Handle negative exponent    if n < 0:        x = 1 / x        n = -n        result = 1.0    current_product = x        # Iterate through bits of n    while n > 0:        # If current bit is 1, multiply result by current_product        if n % 2 == 1:            result *= current_product                # Square the current_product for next iteration        current_product *= current_product                # Move to next bit        n //= 2        return result
```

## Time and Space Complexity
* *Time Complexity**: O(log n)* The algorithm reduces the exponent by half in each recursive call or iteration.* This results in a logarithmic number of operations with respect to n.* *Space Complexity**:* Recursive approach: O(log n) due to the recursion stack* Iterative approach: O(1) as we only use a constant amount of extra space
