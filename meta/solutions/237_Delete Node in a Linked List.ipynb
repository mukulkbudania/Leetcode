## Solution Explanation
This problem has an interesting twist: we don't have access to the head of the linked list, only the node that needs to be deleted. In a typical linked list deletion, we would need the previous node to update its next pointer. However, since we don't have access to the previous node, we need a different approach.The key insight is that we can "delete" the current node by copying the value from the next node into the current node, and then deleting the next node instead. This effectively makes the current node take on the identity of the next node, and then we remove the next node from the list.The algorithm is as follows:1. Copy the value from the next node to the current node2. Update the current node's next pointer to skip over the next node (point to next.next)3. The next node is effectively deleted from the listThis approach works because we're guaranteed that the node to be deleted is not the tail node, so there will always be a next node whose value we can copy.

```python
# Definition for singly-linked list.class ListNode:    def __init__(self, x):        self.val = x        self.next = Noneclass Solution:    def deleteNode(self, node):        """        :type node: ListNode        :rtype: void Do not return anything, modify node in-place instead.        """        # Copy the value from the next node        node.val = node.next.val                # Skip over the next node        node.next = node.next.next
```

## Time and Space Complexity
* *Time Complexity**: O(1) - We perform a constant number of operations regardless of the size of the linked list. We're just copying a value and updating a pointer, both of which are O(1) operations.* *Space Complexity**: O(1) - We don't use any extra space that scales with the input size. We only modify the existing nodes in-place.
