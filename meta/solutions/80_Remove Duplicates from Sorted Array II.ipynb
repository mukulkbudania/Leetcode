## Solution Explanation
This problem asks us to remove duplicates from a sorted array such that each unique element appears at most twice, and to do it in-place.The key insight is that since the array is already sorted, duplicates will be adjacent to each other. We can use a two-pointer approach:1. A slow pointer (`i`) that points to the position where the next valid element should be placed2. A fast pointer (`j`) that scans through the arrayAs we iterate through the array with the fast pointer, we need to decide whether to keep the current element. We should keep it if:* It's different from the element at position `i-1` (meaning it's a new number)* Or if it's the same as the element at position `i-1` but different from the element at position `i-2` (meaning it's the second occurrence of a number)If either condition is met, we place the current element at position `i` and increment `i`.

```python
def removeDuplicates(nums):    # Handle edge cases    if len(nums) <= 2:        return len(nums)        # Initialize the slow pointer (position to place next valid element)    i = 2        # Iterate through the array starting from the third element    for j in range(2, len(nums)):        # If current element is different from the element two positions back in the result array        # it means we can include it (either it's a new number or at most the second duplicate)        if nums[j] != nums[i-2]:            nums[i] = nums[j]            i += 1        return i
```

## Time and Space Complexity
* *Time Complexity**: O(n), where n is the length of the input array. We iterate through the array once with the fast pointer.* *Space Complexity**: O(1), as we only use a constant amount of extra space regardless of the input size. We modify the array in-place without using any additional data structures that scale with input size.
