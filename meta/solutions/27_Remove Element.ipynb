## Solution Explanation
This problem asks us to remove all occurrences of a specific value from an array in-place and return the number of elements that remain.The key insight is that we don't need to actually remove elements (which would be expensive in an array), but rather we need to ensure that the first k elements of the array contain all the elements that are not equal to the target value.I'll use the two-pointer technique:1. Initialize a pointer `k` at the beginning of the array.2. Iterate through the array with another pointer `i`.3. When we find an element that is not equal to the target value, we copy it to the position indicated by `k` and increment `k`.4. At the end, `k` will be the count of elements not equal to the target value.This approach ensures that all elements not equal to the target value are moved to the front of the array, and we return the count of such elements.

```python
def removeElement(nums, val):    """    Remove all occurrences of val in nums in-place and return the new length.        Args:        nums: List[int] - The input array        val: int - The value to remove            Returns:        int - The number of elements in nums which are not equal to val    """    k = 0  # Position to place the next element that's not equal to val        for i in range(len(nums)):        if nums[i] != val:            nums[k] = nums[i]            k += 1        return k
```

## Time and Space Complexity
* *Time Complexity**: O(n), where n is the length of the input array. We iterate through the array once.* *Space Complexity**: O(1), as we're modifying the array in-place and only using a constant amount of extra space regardless of the input size.
