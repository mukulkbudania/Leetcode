## Solution Explanation
This problem asks us to remove duplicates from a sorted array in-place, meaning we need to modify the original array without using extra space for another array.Since the array is already sorted, all duplicates will be adjacent to each other. We can use a two-pointer approach:1. Keep a pointer `i` that points to the position where the next unique element should go2. Iterate through the array with another pointer `j`3. When we find a new unique element (different from the previous one), we place it at position `i` and increment `i`4. Continue until we've processed the entire array5. Return `i` as the count of unique elementsThe key insight is that we only need to move elements when we encounter a new unique value. Since the array is sorted, we can simply compare each element with the previous unique element we've found.

```python
def removeDuplicates(nums):    if not nums:        return 0            # i is the position where the next unique element should go    i = 1        # Iterate through the array starting from the second element    for j in range(1, len(nums)):        # If the current element is different from the previous one        if nums[j] != nums[j-1]:            # Place it at position i            nums[i] = nums[j]            # Move i to the next position            i += 1        return i
```

## Time and Space Complexity
* *Time Complexity**: O(n), where n is the length of the input array. We iterate through the array once.* *Space Complexity**: O(1), as we only use a constant amount of extra space regardless of the input size. We modify the array in-place without using additional data structures.
