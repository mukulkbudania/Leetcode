# 3527. Find the Most Common Response

You are given a 2D string array responses where each responses[i] is an array of strings representing survey responses from the ith day.Return the most common response across all days after removing duplicate responses within each responses[i]. If there is a tie, return the lexicographically smallest response. **Example 1:**Input: responses = [["good","ok","good","ok"],["ok","bad","good","ok","ok"],["good"],["bad"]]Output: "good"Explanation:After removing duplicates within each list, responses = [["good", "ok"], ["ok", "bad", "good"], ["good"], ["bad"]]."good" appears 3 times, "ok" appears 2 times, and "bad" appears 2 times.Return "good" because it has the highest frequency.**Example 2:**Input: responses = [["good","ok","good"],["ok","bad"],["bad","notsure"],["great","good"]]Output: "bad"Explanation:After removing duplicates within each list we have responses = [["good", "ok"], ["ok", "bad"], ["bad", "notsure"], ["great", "good"]]."bad", "good", and "ok" each occur 2 times.The output is "bad" because it is the lexicographically smallest amongst the words with the highest frequency. **Constraints:**1 <= responses.length <= 10001 <= responses[i].length <= 10001 <= responses[i][j].length <= 10responses[i][j] consists of only lowercase English letters

## Solution Explanation
To solve this problem, we need to:1. Remove duplicate responses within each day's responses2. Count the frequency of each unique response across all days3. Find the response with the highest frequency4. If there are multiple responses with the highest frequency, return the lexicographically smallest oneThe approach is straightforward:1. Iterate through each day's responses and convert each day's array to a set to remove duplicates2. Use a Counter to count the frequency of each unique response across all days3. Find the response with the highest frequency4. If there are multiple responses with the highest frequency, return the lexicographically smallest one

```python
from collections import Counterfrom typing import Listdef mostCommonResponse(responses: List[List[str]]) -> str:    # Remove duplicates within each day's responses    unique_responses = [set(day) for day in responses]        # Flatten the list of sets and count frequencies    all_responses = []    for day_set in unique_responses:        all_responses.extend(day_set)        # Count frequencies    counter = Counter(all_responses)        # Find the highest frequency    max_frequency = max(counter.values())        # Find all responses with the highest frequency    most_common = [response for response, freq in counter.items() if freq == max_frequency]        # Return the lexicographically smallest response    return min(most_common)
```

## Time and Space Complexity
* *Time Complexity:*** Converting each day's responses to a set: O(n), where n is the total number of responses across all days* Flattening the list of sets: O(n)* Counting frequencies: O(n)* Finding the highest frequency: O(k), where k is the number of unique responses* Finding all responses with the highest frequency: O(k)* Finding the lexicographically smallest response: O(k)Overall time complexity: O(n), where n is the total number of responses across all days.* *Space Complexity:*** unique_responses: O(n) in the worst case* all_responses: O(n) in the worst case* counter: O(k), where k is the number of unique responses* most_common: O(k) in the worst caseOverall space complexity: O(n), where n is the total number of responses across all days.
