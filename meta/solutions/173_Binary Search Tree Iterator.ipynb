## Solution Explanation
This problem asks us to implement an iterator for in-order traversal of a Binary Search Tree (BST). There are two main approaches:1. **Eager Approach**: Perform a complete in-order traversal during initialization and store all values in an array. Then simply iterate through this array.2. **Lazy Approach**: Use a stack to simulate the recursion of in-order traversal, processing nodes only when needed.I'll implement the lazy approach as it's more space-efficient (O(h) where h is the height of the tree) and follows the follow-up requirement. The key insight is to use a stack to keep track of nodes we need to visit:1. In the constructor, we'll initialize a stack and push all left nodes from the root.2. For `next()`, we pop the top node from the stack, push all left nodes of its right child to the stack, and return the popped node's value.3. For `hasNext()`, we simply check if the stack is empty.This approach ensures O(1) amortized time complexity for both operations and uses O(h) space.

```python
# Definition for a binary tree node.# class TreeNode:#     def __init__(self, val=0, left=None, right=None):#         self.val = val#         self.left = left#         self.right = rightclass BSTIterator:    def __init__(self, root: TreeNode):        # Initialize stack to store nodes        self.stack = []        # Push all left nodes to the stack        self._push_left(root)        def next(self) -> int:        # Pop the next smallest element        node = self.stack.pop()        # If this node has a right child, push all left nodes of the right child        if node.right:            self._push_left(node.right)        # Return the value of the popped node        return node.val        def hasNext(self) -> bool:        # If stack is not empty, there are more elements        return len(self.stack) > 0        def _push_left(self, node):        # Helper method to push all left nodes to the stack        while node:            self.stack.append(node)            node = node.left
```

## Time and Space Complexity
* *Time Complexity:*** Constructor: O(h) where h is the height of the tree, as we push at most h nodes to the stack initially.* next(): Amortized O(1). While a single call might take O(h) time (when we need to push all left nodes of a right child), across n calls to next() for a tree with n nodes, we process each node exactly once, making the amortized cost O(1).* hasNext(): O(1) as we just check if the stack is empty.* *Space Complexity:*** O(h) where h is the height of the tree. In the worst case (a skewed tree), h could be n, but for a balanced tree, h would be log(n).
