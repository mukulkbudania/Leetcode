# 545. Boundary of Binary Tree

The boundary of a binary tree is the concatenation of the root, the left boundary, the leaves ordered from left-to-right, and the reverse order of the right boundary.The left boundary is the set of nodes defined by the following:The root node's left child is in the left boundary. If the root does not have a left child, then the left boundary is empty.If a node is in the left boundary and has a left child, then the left child is in the left boundary.If a node is in the left boundary, has no left child, but has a right child, then the right child is in the left boundary.The leftmost leaf is not in the left boundary.The right boundary is similar to the left boundary, except it is the right side of the root's right subtree. Again, the leaf is not part of the right boundary, and the right boundary is empty if the root does not have a right child.The leaves are nodes that do not have any children. For this problem, the root is not a leaf.Given the root of a binary tree, return the values of its boundary. **Example 1:**Input: root = [1,null,2,3,4]Output: [1,3,4,2]Explanation:- The left boundary is empty because the root does not have a left child.- The right boundary follows the path starting from the root's right child 2 -> 4.  4 is a leaf, so the right boundary is [2].- The leaves from left to right are [3,4].Concatenating everything results in [1] + [] + [3,4] + [2] = [1,3,4,2].**Example 2:**Input: root = [1,2,3,4,5,6,null,null,null,7,8,9,10]Output: [1,2,4,7,8,9,10,6,3]Explanation:- The left boundary follows the path starting from the root's left child 2 -> 4.  4 is a leaf, so the left boundary is [2].- The right boundary follows the path starting from the root's right child 3 -> 6 -> 10.  10 is a leaf, so the right boundary is [3,6], and in reverse order is [6,3].- The leaves from left to right are [4,7,8,9,10].Concatenating everything results in [1] + [2] + [4,7,8,9,10] + [6,3] = [1,2,4,7,8,9,10,6,3]. **Constraints:**The number of nodes in the tree is in the range [1, 104].-1000 <= Node.val <= 1000

## Solution Explanation
To solve this problem, we need to collect the boundary of a binary tree in the following order:1. The root node2. The left boundary (excluding the leftmost leaf)3. All leaf nodes from left to right4. The right boundary in reverse order (excluding the rightmost leaf)The approach involves three separate traversals:1. First, we collect the left boundary by starting from the root's left child and following the left-most path. If there's no left child, we try the right child.2. Second, we collect all leaf nodes using an in-order traversal.3. Finally, we collect the right boundary by starting from the root's right child and following the right-most path, then reverse this list.We need to be careful about a few things:* The root node is handled separately and is not considered a leaf even if it has no children.* Nodes should not be duplicated in the result.* The left and right boundaries exclude the leaves that would be included in the leaf collection.

```python
# Definition for a binary tree node.class TreeNode:    def __init__(self, val=0, left=None, right=None):        self.val = val        self.left = left        self.right = rightclass Solution:    def boundaryOfBinaryTree(self, root: TreeNode) -> list[int]:        if not root:            return []                # If root is the only node        if not root.left and not root.right:            return [root.val]                result = [root.val]  # Start with the root                # Collect left boundary (excluding leaves)        def collect_left_boundary(node):            if not node or (not node.left and not node.right):                return  # Skip if node is None or a leaf                        result.append(node.val)                        if node.left:                collect_left_boundary(node.left)            else:                collect_left_boundary(node.right)                # Collect leaves (in-order traversal)        def collect_leaves(node):            if not node:                return                        collect_leaves(node.left)                        # Check if it's a leaf node            if not node.left and not node.right and node != root:                result.append(node.val)                            collect_leaves(node.right)                # Collect right boundary (excluding leaves) in reverse order        def collect_right_boundary(node):            if not node or (not node.left and not node.right):                return  # Skip if node is None or a leaf                        right_boundary.append(node.val)                        if node.right:                collect_right_boundary(node.right)            else:                collect_right_boundary(node.left)                # Collect all parts of the boundary        if root.left:            collect_left_boundary(root.left)                collect_leaves(root)                right_boundary = []        if root.right:            collect_right_boundary(root.right)                # Add the right boundary in reverse order        result.extend(right_boundary[::-1])                return result
```

## Time and Space Complexity
* *Time Complexity**: O(n), where n is the number of nodes in the tree. We visit each node at most once across all our traversals.* *Space Complexity**: O(h) for the recursion stack, where h is the height of the tree. In the worst case (a skewed tree), this could be O(n). Additionally, we use O(n) space to store the result and the right boundary temporarily. Overall, the space complexity is O(n).
