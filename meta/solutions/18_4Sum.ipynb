# 18. 4Sum

Given an array nums of n integers, return an array of all the unique quadruplets [nums[a], nums[b], nums[c], nums[d]] such that:0 <= a, b, c, d < na, b, c, and d are distinct.nums[a] + nums[b] + nums[c] + nums[d] == targetYou may return the answer in any order. **Example 1:**Input: nums = [1,0,-1,0,-2,2], target = 0Output: [[-2,-1,1,2],[-2,0,0,2],[-1,0,0,1]]**Example 2:**Input: nums = [2,2,2,2,2], target = 8Output: [[2,2,2,2]] **Constraints:**1 <= nums.length <= 200-109 <= nums[i] <= 109-109 <= target <= 109

## Solution Explanation
This problem asks us to find all unique quadruplets in an array that sum up to a given target. This is an extension of the 3Sum problem, which itself is an extension of the 2Sum problem.The approach I'll use is:1. Sort the array first to make it easier to handle duplicates and use two pointers technique2. Use two nested loops to fix the first two elements3. For the remaining two elements, use the two-pointer technique to find pairs that complete the sum to target4. Skip duplicates at each level to ensure unique quadrupletsThe key optimizations:* Early termination: If the smallest possible sum (first 4 elements) exceeds the target, or the largest possible sum (last 4 elements) is less than the target, we can return early* Skip duplicates: After finding a valid quadruplet, skip duplicate elements to avoid duplicate results* Two-pointer technique: For the last two elements, use left and right pointers to efficiently find pairs that sum to the remaining target

```python
def fourSum(nums, target):    nums.sort()    n = len(nums)    result = []        # Early termination    if n < 4:        return result        # If smallest possible sum > target or largest possible sum < target, return empty    if nums[0] + nums[1] + nums[2] + nums[3] > target or nums[n-4] + nums[n-3] + nums[n-2] + nums[n-1] < target:        return result        for i in range(n - 3):        # Skip duplicates for first element        if i > 0 and nums[i] == nums[i - 1]:            continue                    # Early termination for this iteration        if nums[i] + nums[i+1] + nums[i+2] + nums[i+3] > target:            break                    # If using the largest elements still doesn't reach target, skip this i        if nums[i] + nums[n-3] + nums[n-2] + nums[n-1] < target:            continue                    for j in range(i + 1, n - 2):            # Skip duplicates for second element            if j > i + 1 and nums[j] == nums[j - 1]:                continue                            # Early termination for this iteration            if nums[i] + nums[j] + nums[j+1] + nums[j+2] > target:                break                            # If using the largest elements still doesn't reach target, skip this j            if nums[i] + nums[j] + nums[n-2] + nums[n-1] < target:                continue                            # Use two pointers for the remaining two elements            left, right = j + 1, n - 1                        while left < right:                curr_sum = nums[i] + nums[j] + nums[left] + nums[right]                                if curr_sum < target:                    left += 1                elif curr_sum > target:                    right -= 1                else:                    # Found a valid quadruplet                    result.append([nums[i], nums[j], nums[left], nums[right]])                                        # Skip duplicates                    while left < right and nums[left] == nums[left + 1]:                        left += 1                    while left < right and nums[right] == nums[right - 1]:                        right -= 1                                            left += 1                    right -= 1        return result
```

## Time and Space Complexity
* *Time Complexity**: O(n³)* Sorting the array takes O(n log n)* The two nested loops for the first two elements take O(n²)* The two-pointer approach for the last two elements takes O(n) in the worst case* Overall, the time complexity is O(n³), which is dominated by the three nested loops* *Space Complexity**: O(1) excluding the output array* We only use a constant amount of extra space for variables* The sorting is typically done in-place in Python's implementation* The result array doesn't count toward space complexity analysis as it's part of the output
