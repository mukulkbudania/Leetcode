# 12. Integer to Roman

Seven different symbols represent Roman numerals with the following values:Symbol	ValueI	1V	5X	10L	50C	100D	500M	1000Roman numerals are formed by appending the conversions of decimal place values from highest to lowest. Converting a decimal place value into a Roman numeral has the following rules:If the value does not start with 4 or 9, select the symbol of the maximal value that can be subtracted from the input, append that symbol to the result, subtract its value, and convert the remainder to a Roman numeral.If the value starts with 4 or 9 use the subtractive form representing one symbol subtracted from the following symbol, for example, 4 is 1 (I) less than 5 (V): IV and 9 is 1 (I) less than 10 (X): IX. Only the following subtractive forms are used: 4 (IV), 9 (IX), 40 (XL), 90 (XC), 400 (CD) and 900 (CM).Only powers of 10 (I, X, C, M) can be appended consecutively at most 3 times to represent multiples of 10. You cannot append 5 (V), 50 (L), or 500 (D) multiple times. If you need to append a symbol 4 times use the subtractive form.Given an integer, convert it to a Roman numeral. **Example 1:**Input: num = 3749Output: "MMMDCCXLIX"Explanation:3000 = MMM as 1000 (M) + 1000 (M) + 1000 (M) 700 = DCC as 500 (D) + 100 (C) + 100 (C)  40 = XL as 10 (X) less of 50 (L)   9 = IX as 1 (I) less of 10 (X)Note: 49 is not 1 (I) less of 50 (L) because the conversion is based on decimal places**Example 2:**Input: num = 58Output: "LVIII"Explanation:50 = L 8 = VIII**Example 3:**Input: num = 1994Output: "MCMXCIV"Explanation:1000 = M 900 = CM  90 = XC   4 = IV **Constraints:**1 <= num <= 3999

## Solution Explanation
To convert an integer to a Roman numeral, we need to follow the rules specified in the problem:1. We need to handle special cases for 4, 9, 40, 90, 400, and 900 using subtractive notation (IV, IX, XL, XC, CD, CM).2. For other values, we use the additive notation (I, V, X, L, C, D, M).The most straightforward approach is to define a mapping of values to their Roman numeral representations, starting from the largest to the smallest. Then, we can iterate through this mapping and greedily subtract the largest possible value at each step.For example, to convert 3749:1. Start with the largest value ≤ 3749, which is 1000 (M). Subtract 1000 from 3749 and append "M" to the result. Repeat this 3 times to get "MMM" with 749 remaining.2. Next, the largest value ≤ 749 is 500 (D). Subtract 500 from 749 and append "D" to the result. Now we have "MMMD" with 249 remaining.3. Continue this process until the number becomes 0.This approach ensures that we follow the rules of Roman numeral representation.

```python
def intToRoman(num: int) -> str:    # Define the mapping of values to Roman numerals    val_to_roman = [        (1000, "M"), (900, "CM"), (500, "D"), (400, "CD"),        (100, "C"), (90, "XC"), (50, "L"), (40, "XL"),        (10, "X"), (9, "IX"), (5, "V"), (4, "IV"), (1, "I")    ]        result = ""        # Iterate through the mapping and greedily subtract the largest possible value    for value, symbol in val_to_roman:        # While the current value can be subtracted from num        while num >= value:            result += symbol  # Append the corresponding Roman numeral            num -= value      # Subtract the value        return result
```

## Time and Space Complexity
* *Time Complexity**: O(1)* The algorithm iterates through a fixed set of 13 Roman numeral values.* For each value, we perform at most a constant number of operations (since the input is constrained to be at most 3999).* Therefore, the time complexity is constant, O(1).* *Space Complexity**: O(1)* We use a fixed amount of space for the `val_to_roman` mapping.* The output string's length is also bounded by a constant (since the input is at most 3999, the output will have at most a constant number of characters).* Therefore, the space complexity is constant, O(1).
